# Data Analysis
- I will try to explore the relatiosnhip between the **Grad.Rate vs Room.Board**

```python
sns.lmplot(x='Room.Board', y='Grad.Rate', data=df,hue='Private')
```

- I actually don't want the fit lines, so I will try to remove them for now

```python
sns.lmplot(x='Room.Board', y='Grad.Rate', data=df,hue='Private',fit_reg=False,
          palette='coolwarm',height=6,aspect=1)
```

- Next I will explore the dataset further by creating a plot for **F.Undergrad vs Outstate**

```python
sns.lmplot(x='Outstate',y='F.Undergrad',data=df,hue='Private',fit_reg=False,
          height=6,aspect=1)
```

- You can tell from the result above that the tuition fee for the **Private Universities** are higher than the tuition fee for the **Public Universities**.

- Next I will try to use seaborn [sns.FacetGrid](https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.FacetGrid.html) to create stacked histogram showing Out of State Tuition based on the Private column.

```python
g = sns.FacetGrid(df,hue='Private',palette='coolwarm',height=6, aspect=2)
g = g.map(plt.hist,'Outstate',bins=20,alpha=0.7)
```

- Taking a quick look at the histograms, we can again see that the **Out of state** tuition for Private Universities are higher than the tuition for Public Universities 

- Create similar plot for **Grad.Rate**

```python
g = sns.FacetGrid(df,hue='Private',palette='coolwarm',height=6, aspect=2)
g = g.map(plt.hist,'Grad.Rate',bins=20,alpha=0.7)
```

- Taking a quick look at the histograms, we can again see that although both histograms appear to be normally distributed, the **Graduation Rate** for Private Universities were higher than those of Public Universities
