- This project used the in-built scikit learn dataset to demonstrate K-Mean Clustering algorithm.
- The main objective of this project is to identify patterns within the dataset.

## Import data visualisation libraries

```python
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read Dataset

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data
```

```python
data[0].shape
```

- Now lets plot out the blobs

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

- We can observe that there is actually two blobs in the above graph
- Next I will try to grab the second item in that tuple to see the cluster that it belongs to.

```python
data[1]
```

- These are the input labels
- No output labels since this is an unsupervised learning algorithm

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

- Now I will create K-Mean Clustering algorithm 

**K mean cluster = 4**

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

- This returns back the labels it believe to be true for the clusters
- Remember that these are the predicted lables for the clusters which is different from the original labels since this is an unsupervised learning

```python
# using matplotlib 
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6))

ax1.set_title("K Means")
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

- Here I created a subplot of one row with two columns and using sharey=True to show equal axis
- The first set of axis plotted the original data and color the data based on what the algorithm thought the lable should look like
- The second set of axis plotted again the original data but the algorithm colored them with the actually factual correct blob assignment
